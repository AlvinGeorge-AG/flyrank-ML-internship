## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python

import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Total volume of pages
total_pages = len(df)

# Percentage of pages declining
declining_pages = (df['trend_direction'] == 'down').sum()
pct_declining = (declining_pages / total_pages) * 100

# Pages that have not been updated for 180+ days but still get impressions
stale_but_visible = ((df['days_since_last_update'] >= 180) & (df['impressions_90d'] >= 100)).sum()

print(f"1. Total pages in the starter dataset: {total_pages:,}")
print(f"2. Declining pages: {declining_pages:,} ({pct_declining:.1f}% of total inventory)")
print(f"3. Stale pages (no updates for 180+ days) with active search impressions: {stale_but_visible:,}")
```
